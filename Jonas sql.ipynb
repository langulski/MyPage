#### 2.6 qual planta é mais comum a venda de veículos

```python
pd.read_sql_query("""

SELECT  
tb_veiculo.cod_planta, tb_planta.nome_planta,
COUNT(tb_veiculo.cod_planta) AS planta_mais_comum

FROM tb_veiculo

JOIN tb_planta
ON  tb_veiculo.cod_planta = tb_planta.cod_planta
    
    
GROUP BY 
tb_veiculo.cod_planta

ORDER BY 
planta_mais_comum DESC;
""",conn)
```

```python
pd.read_sql_query("""

SELECT  
    tb_veiculo.cod_concessionaria,
    tb_concessionaria.nome_concessionaria,
    tb_concessionaria.estado_concessionaria,
    count(*) AS total_vendas
        
FROM 
   tb_veiculo
   
JOIN  tb_concessionaria
    ON
     tb_veiculo.cod_concessionaria =  tb_concessionaria.cod_concessionaria

GROUP BY 
tb_veiculo.cod_concessionaria


ORDER BY 
    total_vendas DESC;
""",conn)
```
